```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
```

```python
df = pd.DataFrame(columns=['info','loc','price'])
headers = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36 Edg/110.0.1587.63',
    'Host':'wh.lianjia.com',
    'Referer': 'https://wh.lianjia.com/ershoufang/hongshan/pg2rs%E6%B4%AA%E5%B1%B1%E5%8C%BA/'
}

```

```python
for i in range(1,101):
    print(i)
    url = f'https://wh.lianjia.com/ershoufang/hongshan/pg{i}rs%E6%B4%AA%E5%B1%B1%E5%8C%BA/'
    resp = requests.get(url,headers=headers)
    soup = BeautifulSoup(resp.text,'lxml')
    datas = soup.find('ul',class_='sellListContent').find_all('li')
    for data in datas:
        location = data.find('div',class_='flood').text.replace('  ','')
        infomation = data.find('div',class_='houseInfo').text
        priceper = data.find('div',class_='totalPrice totalPrice2').text
        df = df.append([{'info':infomation,'loc':location,'price':priceper}])
    time.sleep(2)
    df.to_excel('house.xlsx')
```
